```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
num_points = 100
x = np.linspace(0, 1, num = num_points)
params = []
means = []
a = 2 #init params
b = 2 #init params
params.append((a,b))
mean = beta.mean(a, b)
means.append(mean)
y0 = beta.pdf(x, a, b) # init distribution
m = 40 # no. of people who liked
n = 10 # no. of people who didn't like 
a += m
b += n
params.append((a,b))
mean = beta.mean(a, b)
means. append(mean)
y1 = beta.pdf(x, a, b) 
#print(y1)
#print(np.mean(y1), np.where(y1 == np.mean(y1))[0])
#print(max(y0), np.where(y0 == max(y0))[0])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(15,4))
ax1.plot(x, y0, "*", label = "(2,2)")
ax2.plot(x, y1, "k--", label = "(2.8, 2)")
ax1.set_title(f'Prior Distribution: Beta{[round(float(value), 2) for value in params[0]]}', fontsize=18)
ax1.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax1.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
ax2.set_title(f'Posterior Distribution: Beta{[round(float(value), 2) for value in params[1]]}', fontsize=18)              
ax2.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax2.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
```

```python
m = 13
n = 17
a += m
b += n
#print (a, b)
params.append((a,b))
y2 = beta.pdf(x, a, b)
mean = beta.mean(a, b)
means.append(mean)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(15,4))
ax1.plot(x, y1, "*", label = f"{params[1]}")
ax2.plot(x, y2, "k--", label = f"{params[2]}")
ax1.set_title(f'Prior Distribution: Beta{[round(float(value), 2) for value in params[1]]}', fontsize=18)
ax1.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax1.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
ax2.set_title(f'Posterior Distribution: Beta{[round(float(value), 2) for value in params[2]]}', fontsize=18)              
ax2.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax2.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
```

```python
m = 70
n = 30
a += m
b += n
params.append((a, b))
y3 = beta.pdf(x, a, b)
mean = beta.mean(a, b)
means.append(mean)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(15,4))
ax1.plot(x, y2, "*", label = f"{params[2]}")
ax2.plot(x, y3, "k--", label = f"{params[3]}")
ax1.set_title(f'Prior Distribution: Beta{[round(float(value), 2) for value in params[2]]}', fontsize=18)
ax1.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax1.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
ax2.set_title(f'Posterior Distribution: Beta{[round(float(value), 2) for value in params[3]]}', fontsize=18)              
ax2.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax2.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
```

```python
fig, ax = plt.subplots(1, 1, sharex = True, figsize = (15, 4))
ax.plot(x, y0, "b--", label = f"{[round(float(value), 2) for value in params[0]]}")
#ax.vlines(x = means[0], ymin = 0, ymax = y1[np.where(y0 == mean(y0))], color = "b", ls = '--')
ax.plot(x, y1, "r--", label = f"{[round(float(value), 2) for value in params[1]]}")
ax.plot(x, y2, "k--", label = f"{[round(float(value), 2) for value in params[2]]}")
ax.plot(x, y3, "g--", label = f"{[round(float(value), 2) for value in params[3]]}")
ax.set_title(f'Distributions', fontsize=18)              
ax.set_xlabel('Value (% of people)', fontsize =16, fontweight='bold', color='gray')
ax.set_ylabel('Probability', fontsize =16, fontweight='bold', color='gray')
ax.legend(fontsize=14, title='Parameters')
```

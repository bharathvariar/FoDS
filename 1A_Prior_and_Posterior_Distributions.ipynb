## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import beta
```

## Implementation

```python
num_points = 100
x = np.linspace(0, 1, num=num_points)
params = []
means = []
a = 2  # init params
b = 2  # init params
params.append((a, b))
mean = beta.mean(a, b)
means.append(mean)
y0 = beta.pdf(x, a, b)  # init distribution
m = 40  # no. of people who liked
n = 10  # no. of people who didn't like
a += m
b += n
params.append((a, b))
mean = beta.mean(a, b)
means.append(mean)
y1 = beta.pdf(x, a, b)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(15, 4))
ax1.plot(x, y0, "b--", label=f"{params[0]}")
ax2.plot(x, y1, "r--", label=f"{params[1]}")
t = ax1.set_title(f"Prior Distribution: Beta{params[0]}", fontsize=18)
xl = ax1.set_xlabel("Value (% of people)", fontsize=16, fontweight="bold", color="gray")
yl = ax1.set_ylabel("Probability Density", fontsize=16, fontweight="bold", color="gray")
t = ax2.set_title(f"Posterior Distribution: Beta{params[1]}", fontsize=18)
xl = ax2.set_xlabel("Value (% of people)", fontsize=16, fontweight="bold", color="gray")
yl = ax2.set_ylabel("Probability Density", fontsize=16, fontweight="bold", color="gray")
```

```python
m = 13
n = 17
a += m
b += n
params.append((a, b))
y2 = beta.pdf(x, a, b)
mean = beta.mean(a, b)
means.append(mean)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharex=True, figsize=(15, 4))
ax1.plot(x, y1, "b--", label=f"{params[1]}")
ax2.plot(x, y2, "r--", label=f"{params[2]}")
t = ax1.set_title(f"Prior Distribution: Beta{params[1]}", fontsize=18)
xl = ax1.set_xlabel("Value (% of people)", fontsize=16, fontweight="bold", color="gray")
yl = ax1.set_ylabel("Probability Density", fontsize=16, fontweight="bold", color="gray")
t = ax2.set_title(f"Posterior Distribution: Beta{params[2]}", fontsize=18)
xl = ax2.set_xlabel("Value (% of people)", fontsize=16, fontweight="bold", color="gray")
yl = ax2.set_ylabel("Probability Density", fontsize=16, fontweight="bold", color="gray")
```
